Use this notebook to upload/list/download MyTorch checkpoints from your HF model repo.

```python
from pathlib import Path
from huggingface_hub import HfApi, hf_hub_download
```

```python
repo_id = 'ShiroOnigami23/MyTorch-MNIST'
checkpoint_path = Path('../checkpoints/mnist_lightweight_mlp.rds')
path_in_repo = 'checkpoints/mnist_lightweight_mlp.rds'
```

```python
api = HfApi()
api.create_repo(repo_id=repo_id, repo_type='model', exist_ok=True)
api.upload_file(
    path_or_fileobj=str(checkpoint_path),
    path_in_repo=path_in_repo,
    repo_id=repo_id,
    repo_type='model'
)
print('Upload complete')
```

```python
api.list_repo_files(repo_id=repo_id, repo_type='model')
```

```python
downloaded = hf_hub_download(repo_id=repo_id, filename=path_in_repo, repo_type='model', local_dir='../checkpoints')
downloaded
```
